```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,5,7,8])
y = np.array([8,13,17,24,26])
```

```python
plt.figure()
plt.scatter(x,y,color="red")
```

```python
h = 3*x+3
print(h)
plt.figure()
plt.plot(x,h,c="b",marker="o")
plt.scatter(x,y,c="r")
```

```python
x.size
```

```python
def cost(y,pred):
  err = pred-y
  kuadrat = err**2
  hasil = kuadrat.sum(axis=0) / (2*y.size)
  return hasil
```

```python
h1 = 3*x+3
print(cost(y,h1))
h2 = 3*x+2
print(cost(y,h2))
h3 = 3*x+2.5
print(cost(y,h3))
h4 = 0.0083*x**4-0.2833*x**3+2.925*x**2-8.1167*x+14.6667
print(cost(y,h4))
```
